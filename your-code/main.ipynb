## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt 

def generar_numeros_uniformes(abajo, techo, contar):
    numeros = np.random.uniform(abajo, techo, contar)
    return numeros
```

```python
fondo = 10
techo = 15
recuento = 100
datos1 = generar_numeros_uniformes(fondo, techo, recuento)
datos1
```

```python
fondo = 10
techo = 60
recuento = 1000
datos2 = generar_numeros_uniformes(fondo, techo, recuento)
datos2
```

```python
def graficar_histograma(datos):
    plt.hist(datos, bins=10)
    plt.xlabel('Valor')
    plt.ylabel('Recuento')
    plt.title('Distribución Uniforme')
    plt.show()
```

```python
graficar_histograma(datos1)
graficar_histograma(datos2)
```

How are the two distributions different?

# your answer below
El primer conjunto de datos estará más concentrado en el rango de 10 a 15, mientras que el segundo conjunto de datos se dispersará en un rango más amplio de 10 a 60.

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here

def generar_numeros_normales(promedio, desviacion, contar):
    numeros = np.random.normal(promedio, desviacion, contar)
    return numeros
```

```python
promedio1 = 10
desviacion1 = 1
recuento1 = 1000
datos1 = generar_numeros_normales(promedio1, desviacion1, recuento1)
datos1
```

```python
promedio2 = 10
desviacion2 = 50
recuento2 = 1000
datos2 = generar_numeros_normales(promedio2, desviacion2, recuento2)
datos2
```

```python
def graficar_histograma(datos):
    plt.hist(datos, bins=50)
    plt.xlabel('Valor')
    plt.ylabel('Recuento')
    plt.title('Distribución Normal')
    plt.show()
```

```python
graficar_histograma(datos1)
graficar_histograma(datos2)
```

How are the two distributions different?

    
# your answer below

El primer histograma presenta una distribución más estrecha y aguda en comparación con el segundo histograma.

La diferencia clave entre las dos distribuciones radica en la desviación estándar. En el primer conjunto de datos, con una desviación estándar de 1, los números aleatorios estarán más concentrados alrededor del valor medio de 10. En el segundo conjunto de datos, con una desviación estándar de 50, los números aleatorios estarán más dispersos a lo largo de un rango más amplio alrededor del valor medio de 10.
